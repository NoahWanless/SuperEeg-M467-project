# For Temporal Data, one option is DCRNN
* DCRNN = Diffusion Convolution + GRU
     * Do diffusion at every time step, + a recurrent temporal update (via the Gated Recurrent Unit)

Normal GNNs do X′=AXW, so update each node with the information from neighboring nodes via a learnable weight matrix (message passing)!
 * Only looks at one hop neighbors

If we do a full connectivity, diffusion matters less, but it does a random walk over k layers of neighbors! Different Ws learn the importance of each layer.

#### This part controls *space*

 

After that, we use a GRU, to control *time*


```python
#Prebuilt in PyG
#Something like this under the hood
class DCRNNCell(nn.Module):
    def __init__(self, in_channels, hidden_dim):
        super().__init__()

        self.conv_x = DiffConv(in_channels, hidden_dim)
        self.conv_h = DiffConv(hidden_dim, hidden_dim)

    def forward(self, x, h_prev, A):
        z = torch.sigmoid(self.conv_x(x, A) + self.conv_h(h_prev, A))
        r = torch.sigmoid(self.conv_x(x, A) + self.conv_h(h_prev, A))

        h_tilde = torch.tanh(
            self.conv_x(x, A) + self.conv_h(r * h_prev, A)
        )

        h = (1 - z) * h_prev + z * h_tilde
        return h
```

DCRNN uses a GRU where every linear transformation is replaced by a graph diffusion operation
* Where the GRU decides what to keep and what to forget from the hidden state
    * The hidden state stores things like:
        * recent signal patterns
        * propagated signals from neighbors
        * longer-term temporal trends
        * if z_t ≈ 1 → trust new signal (x_t)
        * if z_t ≈ 0 → keep old memory (h_{t-1})
(RNN!)
